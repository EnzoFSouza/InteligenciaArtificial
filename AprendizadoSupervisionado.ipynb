```python
"""
Regressão Supervisionada - Concrete Compressive Strength
=========================================================
Modelos: Linear Regression, Decision Tree, Random Forest,
         XGBoost (via GradientBoosting), SVR, MLP (ANN)
Métricas: MAE, MSE, RMSE, R²
"""

# ── Imports ──────────────────────────────────────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ── 1. Carregamento dos dados ────────────────────────────────────────────────
print("=" * 60)
print("  CONCRETE COMPRESSIVE STRENGTH — REGRESSÃO SUPERVISIONADA")
print("=" * 60)

# ── Tenta ucimlrepo; se não disponível, usa réplica sintética ────────────────
from ucimlrepo import fetch_ucirepo
dataset = fetch_ucirepo(id=165)
X = dataset.data.features
y = dataset.data.targets.squeeze()
print("\n[Dataset] UCI Concrete Compressive Strength carregado via ucimlrepo.")

print(f"\nDataset shape  : {X.shape[0]} amostras × {X.shape[1]} features")
print(f"Target         : {y.name}  |  range [{y.min():.2f}, {y.max():.2f}] MPa\n")

# ── 2. Pré-processamento ─────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

# ── 3. Modelos ───────────────────────────────────────────────────────────────
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree":     DecisionTreeRegressor(random_state=42),
    "Random Forest":     RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1),
    "XGBoost (GB)":      GradientBoostingRegressor(n_estimators=300, learning_rate=0.05,
                                                   max_depth=6, random_state=42),
    "SVR":               SVR(kernel="rbf", C=100, epsilon=0.1),
    "MLP (ANN)":         MLPRegressor(hidden_layer_sizes=(128, 64, 32),
                                      activation="relu", max_iter=1000,
                                      random_state=42, early_stopping=True),
}

# Modelos que precisam de dados escalados
SCALED = {"Linear Regression", "SVR", "MLP (ANN)"}

# ── 4. Treino + avaliação ────────────────────────────────────────────────────
results = {}

for name, model in models.items():
    Xtr = X_train_sc if name in SCALED else X_train
    Xte = X_test_sc  if name in SCALED else X_test

    model.fit(Xtr, y_train)
    preds = model.predict(Xte)

    mae  = mean_absolute_error(y_test, preds)
    mse  = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)
    r2   = r2_score(y_test, preds)

    results[name] = {"MAE": mae, "MSE": mse, "RMSE": rmse, "R²": r2, "preds": preds}

    #print(f"{name:<22}  MAE={mae:6.3f}  RMSE={rmse:6.3f}  R²={r2:.4f}")
```

```python
# ── 5. Tabela de resultados ──────────────────────────────────────────────────
results_table = pd.DataFrame(results).T

# Remove a coluna de previsões para exibir apenas métricas
results_table = results_table.drop(columns=["preds"])

# Organiza casas decimais
results_table = results_table.round(4)

print("\n" + "=" * 60)
print("RESULTADOS DOS MODELOS DE REGRESSÃO")
print("=" * 60)

print(results_table)

# Opcional: salvar em CSV
results_table.to_csv("resultados_regressao.csv")

# Opcional: exibir tabela mais bonita
#print("\nTabela formatada:\n")
#print(results_table.to_string())
styled_regression_table = (
    results_table.style
    .highlight_min(subset=["MAE", "MSE", "RMSE"],
                   color="lightgreen")
    .highlight_max(subset=["R²"],
                   color="lightblue")
)

styled_regression_table
```

```python
"""
Classificação Supervisionada - National Poll on Healthy Aging (NPHA)
=====================================================================
Modelos : Logistic Regression, Decision Tree, Random Forest,
          XGBoost (GradientBoosting), SVC, MLP (ANN)
Métricas: Accuracy, Precision, Recall, F1, AUC-ROC, Confusion Matrix
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
from matplotlib.colors import LinearSegmentedColormap
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    ConfusionMatrixDisplay, roc_curve
)

# ── 1. Dados ─────────────────────────────────────────────────────────────────
print("=" * 65)
print("  NATIONAL POLL ON HEALTHY AGING — CLASSIFICAÇÃO SUPERVISIONADA")
print("=" * 65)

from ucimlrepo import fetch_ucirepo
ds = fetch_ucirepo(id=936)
X_raw = ds.data.features.copy()
y_raw = ds.data.targets.squeeze()
print("\n[Dataset] NPHA carregado via ucimlrepo.")

print(f"\nShape   : {X_raw.shape[0]} amostras × {X_raw.shape[1]} features")
print(f"Target  : {y_raw.name}")
print(f"Classes : {sorted(y_raw.dropna().unique())}")
print(f"Dist.   :\n{y_raw.value_counts().sort_index().to_string()}\n")

# ── 2. Pré-processamento ─────────────────────────────────────────────────────
# Codificar target
le = LabelEncoder()
y = le.fit_transform(y_raw.astype(str))
n_classes = len(le.classes_)

# Garantir que X é numérico
X = X_raw.apply(pd.to_numeric, errors="coerce")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Pipeline de pré-processamento (imputer + scaler)
preproc = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("scaler",  StandardScaler()),
])
X_train_sc = preproc.fit_transform(X_train)
X_test_sc  = preproc.transform(X_test)

# ── 3. Modelos ────────────────────────────────────────────────────────────────
models = {
    "Logistic Regression": LogisticRegression(C=10,
                                              solver="lbfgs",
                                              max_iter=3000,
                                              class_weight="balanced",
                                              random_state=42
                                              ),

    "Decision Tree":      DecisionTreeClassifier(max_depth=10,
                                            min_samples_leaf=5,
                                            random_state=42,
                                            class_weight="balanced"
                                            ),

    "Random Forest":       RandomForestClassifier(max_depth=15,
                                                  n_estimators=500,
                                                  min_samples_leaf=1,
                                                  random_state=42,
                                                  max_features="sqrt",
                                                  class_weight="balanced",
                                                  n_jobs=-1
                                                  ),

    "XGBoost (GB)":        GradientBoostingClassifier(n_estimators=500,
                                                      learning_rate=0.01,
                                                      max_depth=3,
                                                      random_state=42
                                                      ),

    "SVC":                 SVC(kernel="rbf",
                               C=100,
                               gamma=0.001,
                               probability=True,
                               random_state=42,
                               class_weight="balanced"
                               ),

    "MLP (ANN)":           MLPClassifier(hidden_layer_sizes=(64, 32),
                                         solver="adam",
                                         activation="relu",
                                         alpha=0.001,
                                         learning_rate_init=0.001,
                                         max_iter=3000,
                                         random_state=42,
                                         early_stopping=True
                                         ),
}

SCALED = {"Logistic Regression", "SVC", "MLP (ANN)"}

# ── 4. Treino + avaliação ─────────────────────────────────────────────────────
avg = "macro"
results = {}

print(f"\n{'Modelo':<22}  {'Acc':>6}  {'Prec':>6}  {'Rec':>6}  {'F1':>6}  {'AUC':>6}")
print("-" * 60)

for name, model in models.items():
    Xtr = X_train_sc if name in SCALED else X_train_sc   # todos escalados pelo pipeline
    Xte = X_test_sc

    model.fit(Xtr, y_train)
    preds = model.predict(Xte)

    if hasattr(model, "predict_proba"):
        proba = model.predict_proba(Xte)
    else:
        proba = None

    acc   = accuracy_score(y_test, preds)
    prec  = precision_score(y_test, preds, average=avg, zero_division=0)
    rec   = recall_score(y_test, preds, average=avg, zero_division=0)
    f1    = f1_score(y_test, preds, average=avg, zero_division=0)
    try:
        if n_classes == 2:
            auc = roc_auc_score(y_test, proba[:,1] if proba is not None else preds)
        else:
            auc = roc_auc_score(y_test, proba, multi_class="ovr", average=avg) if proba is not None else np.nan
    except Exception:
        auc = np.nan

    cm = confusion_matrix(y_test, preds)
    results[name] = dict(acc=acc, prec=prec, rec=rec, f1=f1, auc=auc,
                          cm=cm, preds=preds, proba=proba)

    auc_str = f"{auc:.4f}" if not np.isnan(auc) else "  N/A "
    print(f"{name:<22}  {acc:.4f}  {prec:.4f}  {rec:.4f}  {f1:.4f}  {auc_str}")
```

```python
# ── 5. Tabela de resultados ──────────────────────────────────────────────────
results_table = pd.DataFrame(results).T

# Mantém apenas métricas
results_table = results_table[["acc", "prec", "rec", "f1", "auc"]]

# Renomeia colunas
results_table.columns = ["Accuracy", "Precision", "Recall", "F1-Score", "AUC"]

# Arredondamento
results_table = results_table.round(4)

print("\n" + "=" * 70)
print("RESULTADOS DOS MODELOS DE CLASSIFICAÇÃO")
print("=" * 70)

print(results_table)

# Opcional: salvar em CSV
results_table.to_csv("resultados_classificacao.csv")

# ── 6. Matrizes de confusão em porcentagem ───────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.ravel()

for ax, (name, data) in zip(axes, results.items()):

    cm = data["cm"]

    # Converter para porcentagem por linha
    cm_percent = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis]
    cm_percent *= 100

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm_percent,
        display_labels=le.classes_
    )

    disp.plot(
        ax=ax,
        cmap="Blues",
        colorbar=False,
        values_format=".1f"
    )

    ax.set_title(name, fontsize=11)

plt.suptitle("Matrizes de Confusão (%)", fontsize=18)
plt.tight_layout()
plt.show()

# Destacar melhores métricas
styled_classification_table = (
    results_table.style
    .highlight_max(subset=["Accuracy", "Precision", "Recall", "F1-Score", "AUC"],
                   color="lightgreen")
)

styled_classification_table
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Binariza classes
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))

plt.figure(figsize=(10,7))

for name, data in results.items():

    proba = data["proba"]

    if proba is None:
        continue

    # ROC macro-average OvR
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Média simples
    mean_auc = np.mean(list(roc_auc.values()))

    # Plot classe 0 só para simplificar visual
    plt.plot(
        fpr[0],
        tpr[0],
        label=f"{name} (AUC={mean_auc:.3f})"
    )

plt.plot([0,1], [0,1], "k--")

plt.xlabel("Taxa de falsos positivos")
plt.ylabel("Taxa de verdadeiros positivos")
plt.title("ROC Curves - Classificação")
plt.legend()
plt.grid(True)
plt.show()
```
